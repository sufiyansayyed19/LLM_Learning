## And Ollama also gives an OpenAI compatible endpoint

...and it's on your local machine!

If the next cell doesn't print "Ollama is running" then please open a terminal and run `ollama serve`

```python
# 1. Install Ollama (Linux command)
!curl -fsSL https://ollama.com/install.sh | sh

# 2. Start the Ollama Server in the background
import subprocess
import time
import requests

# Run "ollama serve" as a detached process
# Standard procedure for running background services in Notebooks
process = subprocess.Popen(["ollama", "serve"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

print("Starting Ollama server...")
time.sleep(5)  # Give the server a few seconds to initialize

# 3. Now run the code you wanted to run
try:
    response = requests.get("http://localhost:11434")
    print(f"\nStatus Code: {response.status_code}")
    print(f"Content: {response.content.decode('utf-8')}") # Decoded for readability
except Exception as e:
    print(f"Connection failed: {e}")
```

### Download llama3.2 from meta

Change this to llama3.2:1b if your computer is smaller.

Don't use llama3.3 or llama4! They are too big for your computer..

```python
from openai import OpenAI

# Standard Naming: Use 'client' instead of 'ollama'
# This creates a reusable client instance pointing to your local server
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"  # Ollama requires a key string, but doesn't validate it. 'ollama' is standard.
)

```

```python
# ----------------------------
# 1. Run Llama 3.2
# ----------------------------
print("ü§ñ Generating with Llama 3.2...")
response_llama = client.chat.completions.create(
    model="llama3.2",
    messages=[{"role": "user", "content": "Tell me a fun fact"}]
)

print(f"Llama says:\n{response_llama.choices[0].message.content}\n")


```

```python
# ----------------------------
# 2. Run DeepSeek R1 (Distilled)
# ----------------------------
# We pull the model using the magic command (!) inside the cell
print("‚¨áÔ∏è  Pulling DeepSeek R1 (this might take a minute)...")
!ollama pull deepseek-r1:1.5b

print("üß† Generating with DeepSeek R1...")
response_deepseek = client.chat.completions.create(
    model="deepseek-r1:1.5b",
    messages=[{"role": "user", "content": "Tell me a fun fact"}]
)

print(f"DeepSeek says:\n{response_deepseek.choices[0].message.content}")
```

## Recap on installation of Ollama

Simply visit [ollama.com](https://ollama.com) and install!

Once complete, the ollama server should already be running locally.  
If you visit:  
[http://localhost:11434/](http://localhost:11434/)

You should see the message `Ollama is running`.  

If not, bring up a new Terminal (Mac) or Powershell (Windows) and enter `ollama serve`  
And in another Terminal (Mac) or Powershell (Windows), enter `ollama pull llama3.2`  
Then try [http://localhost:11434/](http://localhost:11434/) again.

If Ollama is slow on your machine, try using `llama3.2:1b` as an alternative. Run `ollama pull llama3.2:1b` from a Terminal or Powershell, and change the code from `MODEL = "llama3.2"` to `MODEL = "llama3.2:1b"`
